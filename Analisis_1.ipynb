```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Analisis exploratorio

```python
loc_2019 = pd.read_csv("1_analisis_eg/local_sin_medidas.csv",sep=";",encoding = 'utf_16_le')
loc_2020 = pd.read_csv("1_analisis_eg/local_con_medidas.csv",sep=";",encoding = 'utf-16_le')
```

```python
# hay valores nulos en ambos años pero son pocas muestras
#loc_2019['TOU'].value_counts()
loc_2020['TOU'].value_counts() # hay valores nulos en ambos años pero son pocas muestras
```

```python
#eliminar observaciones nulas
loc_2019 = loc_2019[loc_2019["TOU"]!=" "] 
loc_2020 = loc_2020[loc_2020["TOU"]!=" "]
```

```python
def format_df(df):  # dar formato para analisis posterior
    df["Data"]              = pd.to_datetime(df["Data"],format='%d/%m/%Y %H:%M',dayfirst=True)
    df["Demanda Activa"]    = df["Demanda Activa"].astype('str')
    df["Demanda Reactiva"]  = df["Demanda Reactiva"].astype('str')
    df["Unidad"]            = df["Unidad"].astype('str')
    df["Unidad.1"]          = df["Unidad.1"].astype('str')
    df["TOU"]               = df["TOU"].astype('str')
    df["Demanda Activa"]    = df["Demanda Activa"].apply(lambda x: x.replace(',','.'))
    df["Demanda Reactiva"]  = df["Demanda Reactiva"].apply(lambda x: x.replace(',','.'))
    df["Demanda Activa"]    = df["Demanda Activa"].astype('float')
    df["Demanda Reactiva"]  = df["Demanda Reactiva"].astype('float')
    df["Potencia aparente"] = np.sqrt(df["Demanda Activa"]**2 + df["Demanda Reactiva"]**2)
    # columnas timestamp
    #df["day"]     = df["Data"].apply(lambda x : x.day)
    #df["month"]   = df["Data"].apply(lambda x : x.month)
    #df["year"]    = df["Data"].apply(lambda x : x.year)
    #df["hour"]    = df["Data"].apply(lambda x : x.hour)
    #df["minute"]  = df["Data"].apply(lambda x : x.minute)
    
    
def plot_pot_ap(df,pprom,title):
    plt.figure(figsize=(16,9))
    plt.title(title)
    plt.plot(df["Data"],df["Potencia aparente"],alpha=0.2,color='blue',label="Potecia aparente [KW]")
    plt.plot(df["Data"],pprom*np.ones(df["Data"].size),color='blue',label="Potencia aparente promedio: "+str(pprom)+" [KW]")
    plt.legend()
    plt.xlabel("Dias")
    plt.ylabel("Potencia")
    
def plot_pot_act_rea(df,p_act,p_rea,title):
    plt.figure(figsize=(16,9))
    plt.title(title)
    plt.plot(df["Data"],df["Demanda Activa"],alpha=0.2,color="green",label="Potencia Activa [KW]")
    plt.plot(df["Data"],df["Demanda Reactiva"],alpha=0.2,color="red",label="Potencia Activa [KVAR]")
    plt.plot(df["Data"],p_act*np.ones(df["Data"].size),color='green',label="Potencia Ativa promedio: "+str(p_act)+" [KW]")
    plt.plot(df["Data"],p_rea*np.ones(df["Data"].size),color='red',label="Potencia Reactiva promedio: "+str(p_rea)+" [KVAR]")
    plt.legend()
    plt.xlabel("Dias")
    plt.ylabel("Potencia")
```

```python
format_df(loc_2019)
format_df(loc_2020)
```

```python
pap_prom_2019 = round(loc_2019["Potencia aparente"].mean(),3)
pap_prom_2020 = round(loc_2020["Potencia aparente"].mean(),3)
print("Potencia promedio aparente 2019: "+str(pap_prom_2019)+ " [KW]")
print("Potencia promedio aparente 2019: "+str(pap_prom_2020)+ " [KW]")
```

```python
plot_pot_ap(loc_2019,pap_prom_2019,"Potencia aparente [KW] año 2019 (AGOSTO)")
plot_pot_ap(loc_2020,pap_prom_2020,"Potencia aparente [KW] año 2020 (AGOSTO)")
```

```python
kw_2019   = round(loc_2019["Demanda Activa"].mean(),3)
kvar_2019 = round(loc_2019["Demanda Reactiva"].mean(),3)
print("KW  2019 : " +str(kw_2019))
print("KVAR 2019: "+str(kvar_2019))
print("*************************")
kw_2020   = round(loc_2020["Demanda Activa"].mean(),3)
kvar_2020 = round(loc_2020["Demanda Reactiva"].mean(),3)
print("KW  2020 : " +str(kw_2020))
print("KVAR 2020: "+str(kvar_2020))
```

```python
plot_pot_act_rea(loc_2019,kw_2019,kvar_2019,"Potencia Activa [KW] y Reactiva [KVAR], año 2019 (AGOSTO)")
plot_pot_act_rea(loc_2020,kw_2020,kvar_2020,"Potencia Activa [KW] y Reactiva [KVAR], año 2020 (AGOSTO)")
```

```python
difkw = round(kw_2020-kw_2019,3)
print("Diferencia potencia activa: "+str(difkw)+" [KW]")
varkw = round(difkw/kw_2019,3)
print("Variacion (%): " + str(varkw*100) +" %")
print("***********************")
difkvar = round(kvar_2020-kvar_2019,3)
print("Diferencia potencia reactiva: "+str(difkvar)+" [KVAR]")
varkvar = round(difkvar/kvar_2019,2)
print("Variacion (%): " + str(varkvar*100) +" %")
print("***********************")
fp_2019 = round(kw_2019/pap_prom_2019,3)
fp_2020 = round(kw_2020/pap_prom_2020,3)
var_fp = round((abs(fp_2020-fp_2019)/fp_2019)*100,2)
print("Eficiencia 2019 (FP) : " + str(fp_2019))
print("Eficiencia 2020 (FP): "  + str(fp_2020))
print("Variacion Eficiencia (%): "+str(var_fp)+" %")
```

# Analisis horas punta y no punta

```python
loc_nhp_2019 = loc_2019[loc_2019["TOU"]=='Fuera de Punta ']
loc_hp_2019  = loc_2019[loc_2019["TOU"]=='Punta ']

loc_nhp_2020 = loc_2020[loc_2020["TOU"]=='Fuera de Punta ']
loc_hp_2020  = loc_2020[loc_2020["TOU"]=='Punta ']
```

```python
print("año 2019")
print("*************************")
kw_nhp_2019   = round(loc_nhp_2019["Demanda Activa"].mean(),2)
kvar_nhp_2019 = round(loc_nhp_2019["Demanda Reactiva"].mean(),2)
print("KW hora no punta 2019 : " +str(kw_nhp_2019))
print("KVAR hora no punta 2019: "+str(kvar_nhp_2019))
print("*************************")
kw_hp_2019   = round(loc_hp_2019["Demanda Activa"].mean(),2)
kvar_hp_2019 = round(loc_hp_2019["Demanda Reactiva"].mean(),2)
print("KW hora punta  2019 : " +str(kw_hp_2019))
print("KVAR hora punta 2019: "+str(kvar_hp_2019))
print("año 2020")
print("*************************")

kw_nhp_2020   = round(loc_nhp_2020["Demanda Activa"].mean(),2)
kvar_nhp_2020 = round(loc_nhp_2020["Demanda Reactiva"].mean(),2)
print("KW hora no punta 2020 : " +str(kw_nhp_2020))
print("KVAR hora no punta 2020: "+str(kvar_nhp_2020))
print("*************************")
kw_hp_2020   = round(loc_hp_2020["Demanda Activa"].mean(),2)
kvar_hp_2020 = round(loc_hp_2020["Demanda Reactiva"].mean(),2)
print("KW hora punta  2020 : " +str(kw_hp_2020))
print("KVAR hora punta 2020: "+str(kvar_hp_2020))
print("*************************")
kw_hp_diff   = round(kw_hp_2019-kw_hp_2020,2)
kvar_hp_diff = kvar_hp_2019-kvar_hp_2020

kw_nhp_diff   = round(kw_nhp_2019-kw_nhp_2020,2)
kvar_nhp_diff = round(kvar_nhp_2019-kvar_nhp_2020,3)

print("Variacion año 2019-2020")
print("*************************")
print("Diferencia KW hora punta: "+str(kw_hp_diff))
print("Diferencia KVAR hora punta: "+str(kvar_hp_diff))
#print("*************************")
print("Diferencia KW hora NO punta: "+str(kw_nhp_diff))
print("Diferencia KVAR hora NO punta: "+str(kvar_nhp_diff))
```
